```python
x=ratings['user_id'].value_counts()>200
y=x[x].index
ratings=ratings[ratings['user_id'].isin(y)]
ratings.shape
```

```python
ratings.head(4)
```

```python
rb=ratings.merge(books,on='ISBN')
```

```python
number_rating=rb.groupby('title')['rating'].count().reset_index()
number_rating
```

```python
number_rating.rename(columns={'rating':'no_of_ratings'},inplace=True)
```

```python
number_rating.head(3)
```

```python
final_rating=rb.merge(number_rating,on='title')
final_rating.head(3)
```

```python
final_rating=final_rating[final_rating['no_of_ratings']>=50]
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.duplicated().sum()
```

```python
final_rating.shape
```

```python
book_pivot=final_rating.pivot_table(columns='user_id',index='title',values='rating')
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
print(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distances,suggestions=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distances
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```

```python
np.where(book_pivot.index=="Animal Farm")[0][0]
```

```python
def recommend_book(book_name):
    id= np.where(book_pivot.index==book_name)[0][0]
    distances,suggestions=model.kneighbors(book_pivot.iloc[id,:].values.reshape(1,-1),n_neighbors=6)
    for i in range(len(suggestions)):
        print(book_pivot.index[suggestions[i]])
```

```python
recommend_book('The Cradle Will Fall')
```
